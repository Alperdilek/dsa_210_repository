```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("final_periodic_workout_data_23Mar_23Apr.csv")
df = df.dropna(subset=["Exercise"])

sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(edgecolor=None, data=df, x="Weight Load (kg)", y="Caloric Expenditure (kcal)", hue="Exercise")
plt.title("Scatter Plot: Weight vs Caloric Expenditure")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="Exercise", y="Caloric Expenditure (kcal)", palette="Set2")
plt.title("Boxplot: Exercise vs Caloric Expenditure")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Date", y="Heart Rate (bpm)", hue="Exercise", marker="o")
plt.xticks(rotation=45)
plt.title("Heart Rate Trend Over Time")
plt.show()
```

```python
plt.figure(figsize=(8,6))
correlation = df[["Weight Load (kg)", "Repetition Count", "Total Volume", "Heart Rate (bpm)", "Caloric Expenditure (kcal)"]].corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df, x="Weight Load (kg)", y="Caloric Expenditure (kcal)", scatter_kws={'alpha':0.5})
plt.title("Regression: Weight vs Caloric Expenditure")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(edgecolor=None, data=df, x="Total Volume", y="Caloric Expenditure (kcal)", hue="Exercise")
plt.title("Scatter Plot: Total Volume vs Caloric Expenditure")
plt.show()
```
